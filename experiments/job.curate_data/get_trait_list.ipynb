```python
trait_list = sorted(set(df_gs['TRAIT']) & set(df_sumstats.index))
df_info_full = df_sumstats.loc[trait_list]
df_info_full['Code'] = [dic_code[x][0] for x in df_info_full['Trait_Identifier']]
df_info_full['Category'] = [dic_code[x][2] for x in df_info_full['Trait_Identifier']]
df_info_full['INCLUDE'] = [dic_code[x][1] for x in df_info_full['Trait_Identifier']]
df_info_full = df_info_full[['Trait Name', 'Trait_Identifier', 'Code', 'Category', 'INCLUDE', 'Reference', 'N', 'M']]
df_info_full = df_info_full.join(df_hsq[['h2g', 'sd', 'Z', 'Intercept']])
df_info_full = df_info_full.join(df_me[['log10 Me Common']])
df_info_full['log10 Me Common'].fillna('', inplace=True)
df_info_full.index = df_info_full['Code']
df_info_full.sort_values(by=['Category', 'Trait_Identifier'], inplace=True, ascending=True)
df_info_full.to_csv(OUT_FILE_FULL, sep='\t', index=False)

df_info = df_info_full.loc[df_info_full['INCLUDE']=='Y'].copy()
df_info.drop('INCLUDE', axis=1, inplace=True)
df_info.index = df_info['Trait_Identifier']
df_info.loc[['UKB_460K' in x for x in df_info['Trait_Identifier']], 'Reference'] = 'UK Biobank'
df_info['Trait Name'] = [dic_trait_name[x] if x in dic_trait_name else df_info.loc[x, 'Trait Name']
                         for x in df_info['Trait_Identifier']]
df_info['Trait Name'] = [x.split('(U')[0].strip() for x in df_info['Trait Name']]
df_info.to_csv(OUT_FILE, sep='\t', index=False)
print('# n_trait=%d/%d'%(df_info.shape[0], df_info_full.shape[0]))
print('# Avg. N=%0.1f'%(df_info['N'].mean()))
```
